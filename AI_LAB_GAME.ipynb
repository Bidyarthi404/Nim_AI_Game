```python
def computer_chooses_move(n, m):
    best_value = float('-inf')
    best_move = 1

    for i in range(1, min(n, m) + 1):
        value = minimax(n - i, m, False)
        if value > best_value:
            best_value = value
            best_move = i

    return best_move

def user_chooses_move(n, m):
    while True:
        global global_pieces_taken
        global_pieces_taken = int(input("\nHow many pieces will you take? "))
        if global_pieces_taken <= m and global_pieces_taken <= n and global_pieces_taken > 0:
            return global_pieces_taken
        else:
            print("\nOops! Unexpected move. Try again!")

def minimax(n, m, is_maximizing):
    if n == 0:
        return -1 if is_maximizing else 1

    if is_maximizing:
        best_value = float('-inf')
        for i in range(1, min(n, m) + 1):
            value = minimax(n - i, m, False)
            best_value = max(best_value, value)
        return best_value
    else:
        best_value = float('inf')
        for i in range(1, min(n, m) + 1):
            value = minimax(n - i, m, True)
            best_value = min(best_value, value)
        return best_value

def match():
    # Player name
    pName = input("\nEnter your name: ")
    print("\nWelcome to AI Lab, ", pName, ", Choose the type of game you want to play:")
    game_mode = input("\n1 - Single match \n2 - Best of 3 championship\n")
    if game_mode == '1':
        print("\nYou chose a Single match")
    elif game_mode == '2':
        print("\nYou chose a Best of 3 championship.")
        computer = 0
        user = 0
        counter = 1
    else:
        print("\nUnexpected value. Please, choose from one of the two")
        return False

    while True:
        if game_mode == '2' and counter == 4:
            # checking if the championship is over
            print("\n**** End of the championship! ****")
            print("\nScoreboard:", pName, user, "X", computer, "Maruf Sir")
            break
        elif game_mode == '2':
            # displaying the round number
            print("\n**** Round", counter, "****")

        n = int(input("\nHow many pieces? "))
        m = int(input("Maximum of pieces per move? "))
        if m < 1 or n < 1:
            print("\nInvalid value. Choose a positive integer for both variables.")
            continue

        if n % (m + 1) == 0:
            # if n is multiple of (m+1), the user must begin so that the computer wins
            c = 0
            print("\nYou begin!")
            while True:
                c += 1
                if c % 2 == 0:
                    if computer_chooses_move(n, m) < 2:
                        print("\nMaruf Sir took one piece.")
                    else:
                        print("\nMaruf Sir took", computer_chooses_move(n, m), "pieces.")
                    n = n - computer_chooses_move(n, m)
                    if n == 0:
                        if game_mode == '1':
                            # if the game mode is an isolated match, there'll be one round only
                            print("\nEnd of the game. Maruf Sir wins!")
                            print("Thank you for playing!")
                            return True
                        elif game_mode == '2':
                            # if the game mode is a championship, it'll go to the next round
                            print("\nEnd of match. Maruf Sir wins!")
                            computer += 1
                            counter += 1
                            break
                    elif n < 2:
                        print("Now there is only one piece on the board.")
                    else:
                        print("Now there are", n, "pieces on the board.")

                else:
                    if user_chooses_move(n, m) < 2:
                        print("\n", pName, " took one piece.")
                    else:
                        print("\n", pName, " took", global_pieces_taken, "pieces.")
                    n = n - global_pieces_taken
                    if n == 0:
                        return "THE END"
                    elif n < 2:
                        print("Now there is only one piece on the board.")
                    else:
                        print("Now there are", n, "pieces on the board.")
        else:
            # if n isn't multiple of (m+1), the computer must begin so that it wins
            c = 0
            print("\nMaruf Sir begins!")
            while True:
                c += 1
                if c % 2 == 0:
                    if user_chooses_move(n, m) < 2:
                        print("\n", pName, " took one piece.")
                    else:
                        print("\n", pName, " took", global_pieces_taken, "pieces.")
                    n = n - global_pieces_taken
                    if n == 0:
                        return "END OF THE GAME>>>>USER WINS"
                    elif n < 2:
                        print("Now there is only one piece on the board.")
                    else:
                        print("Now there are", n, "pieces on the board.")

                else:
                    # Use the minimax function to determine the computer's move
                    move = computer_chooses_move(n, m)
                    if move < 2:
                        print("\nMaruf Sir took one piece.")
                    else:
                        print("\nMaruf Sir took", move, "pieces.")
                    n = n - move
                    if n == 0:
                        if game_mode == '1':
                            print("\nEnd of the game. Maruf Sir wins!")
                            print("Thank you for playing!")
                            return "THE END"
                        elif game_mode == '2':
                            print("\nEnd of round. Maruf Sir wins!")
                            computer += 1
                            counter += 1
                            break
                    elif n < 2:
                        print("Now there is only one piece on the board.")
                    else:
                        print("Now there are", n, "pieces on the board.")

match()
```
